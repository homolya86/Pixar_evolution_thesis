```python
# A beolvasott szövegkönyvben a [ karakterek elé beszúr egy : -ot.

def modify_script(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        print("Az összes sor a szövegkönyvben: ", len(lines))
    modified_lines = []
    x = 0
    
    for line in lines:
        if line.startswith('['):
            modified_lines.append(':[' + line[1:])
            x = x+1
        else:
            modified_lines.append(line)

    with open(output_file, 'w', encoding='utf-8') as file:
        file.writelines(modified_lines)
    print("Összesen ", x, " db módosítás lett végezve.")

# Fájlnevek
input_file = '/Users/markuslorinc/Downloads/Cars 3 (2017)/Cars 3 (2017)-transcript.txt'  # Az input fájl neve, amiben a szövegkönyv található
output_file = '/Users/markuslorinc/Downloads/Cars 3 (2017)/Cars_2017_modified.txt'  # A módosított szövegkönyv kimeneti fájlja

# Módosítás és kimeneti fájl létrehozása
modify_script(input_file, output_file)

print("A módosított szövegkönyv elkészült és el lett mentve.")
```

```python
# Mondattípus megszámlálása, egyesszám-többesszám mondatok megszámlálása, szavak gyakoriságának számolása.

import os
import xlwt
from collections import Counter
import re

# Mappa elérési útvonala
folder_path = "/Users/markuslorinc/Downloads/Toy Story 4 (2019)/Characters"

# Stop szavak fájl elérési útvonala
stopwords_file = "/Users/markuslorinc/Downloads/stopwords.txt"

# Egyes számú névmások fájl elérési útvonala
singular_file = "/Users/markuslorinc/Downloads/Singular first-person pronouns.txt"

# Többes számú névmások fájl elérési útvonala
plural_file = "/Users/markuslorinc/Downloads/Plural first-person pronouns.txt"

# Függvény a szavak számolására, kiszűrve a stopszavakat
def count_words(text, stopwords):
    words = re.findall(r'\b\w+\b', text.lower())
    words = [word for word in words if word not in stopwords]
    return len(set(words))

# Függvény a mondattípusok számolására
def count_sentences(text):
    dot_count = 0
    question_mark_count = 0
    exclamation_mark_count = 0
    previous_char = None

    for char in text:
        if char == '.':
            if previous_char != '.':
                dot_count += 1
        elif char == '?':
            if previous_char != '?':
                question_mark_count += 1
        elif char == '!':
            if previous_char != '!':
                exclamation_mark_count += 1
        previous_char = char

    return dot_count, question_mark_count, exclamation_mark_count


# Fájl tartalmának beolvasása listaként
def read_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        return file.read()

# Mappában lévő összes txt fájl elérési útvonalának összegyűjtése
txt_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith('.txt')]

# Stop szavak beolvasása
stopwords = set(read_file(stopwords_file).split())

# Egyes számú névmások beolvasása
singular_words = set(read_file(singular_file).split())

# Többes számú névmások beolvasása
plural_words = set(read_file(plural_file).split())


# Excel munkafájl létrehozása vagy megnyitása
workbook = xlwt.Workbook()

# Mappa neve
folder_name = os.path.basename(folder_path)


for file_path in txt_files:
    character_name = os.path.splitext(os.path.basename(file_path))[0]  # Karakter neve a fájlnév alapján

    # Fájl tartalmának beolvasása
    text = read_file(file_path)

    # Mondattípusok számolása
    sentences_counts = count_sentences(text)

    # Egyes számú névmások számolása
    singular_count = sum(1 for word in re.findall(r'\b\w+\b', text) if word in singular_words)

    # Többes számú névmások számolása
    plural_count = sum(1 for word in re.findall(r'\b\w+\b', text) if word in plural_words)

    # Különböző szavak számolása
    unique_words_count = count_words(text, stopwords)

    # Excel worksheet létrehozása
    worksheet = workbook.add_sheet(character_name)

    # Adatok írása az Excel táblázatba csak egyszer, ha még nem létezik a munkalap
    worksheet.write(0, 0, "Kijelentő mondatok")
    worksheet.write(0, 1, "Kérdő mondatok")
    worksheet.write(0, 2, "Felkiáltó mondatok")
    worksheet.write(0, 3, "Egyes számú névmások száma")
    worksheet.write(0, 4, "Többes számú névmások száma")
    worksheet.write(0, 5, "Különböző szavak száma")

    # Adatok írása az Excel táblázatba a karakterhez tartozó sorba
    row = len(worksheet.rows)
    worksheet.write(row, 0, sentences_counts[0])
    worksheet.write(row, 1, sentences_counts[1])
    worksheet.write(row, 2, sentences_counts[2])
    worksheet.write(row, 3, singular_count)
    worksheet.write(row, 4, plural_count)
    worksheet.write(row, 5, unique_words_count)
    
    print('Az alábbi karakter: ', character_name, ' elemzése elkészült.')

# Excel fájl mentése
workbook.save(os.path.join(folder_path, f"{folder_name}.xls"))
```
